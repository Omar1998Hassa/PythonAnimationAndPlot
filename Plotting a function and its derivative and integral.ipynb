```python
import numpy as np
from cmath import sin
import matplotlib.pyplot as plt
from scipy import integrate
```

# ploting function sin(x)*e^(-0.2*x) with its integration and differentiation

```python
func  = lambda x : np.sin(x) * np.exp(-0.2*x)
exp_func  = lambda x : x**2
def function(start,end, numOfpoints, func ):
    return func(np.linspace(start,end,numOfpoints))
```

```python
def diff(x,y): # get derivative of y with respect to x 
    diffy= y[1:]- y[:-1]
    diffx=x[1:]-x[:-1]
    return diffy/diffx
```

```python

y= function(0,10, 10000,func)
x = np.linspace(0,10,10000)


print(diff(x,y))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y)
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,integrate.cumtrapz(y,x, initial=0)) # integration of y with respect to x
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x[:-1],diff(x,y))# differentiation of y with respect to x
```

```python
def zerocrossing(y):
    yshifted = y[1:]
    multArray = yshifted*y[:-1]
    return multArray <=0
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y)
plt.plot(x,integrate.cumtrapz(y,x, initial=0))
plt.plot(x[:-1],diff(x,y))
plt.legend(["function","integration","differentiaion"])
plt.plot((x[:-1])[zerocrossing(y)],(y[:-1])[zerocrossing(y)],linestyle='', marker='o', color='red')
plt.plot((x[:-2])[zerocrossing(diff(x,y)) ],(y[:-2])[zerocrossing(diff(x,y))],linestyle='', marker='o', color='black')
plt.show()


```
